```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target']=iris.target
df['flower']=df['target'].apply(lambda x: iris.target_names[x])
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC(C=30, kernel='rbf', gamma='auto')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

Using the K Fold Cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(C=30, kernel='rbf', gamma='auto'),iris.data, iris.target, cv=5)
```

```python
cross_val_score(SVC(C=20, kernel='rbf', gamma='auto'),iris.data, iris.target, cv=5)
```

```python
cross_val_score(SVC(C=1, kernel='rbf', gamma='auto'),iris.data, iris.target, cv=5)
```

```python
cross_val_score(SVC(C=30, kernel='linear', gamma='auto'),iris.data, iris.target, cv=5)
```

```python
cross_val_score(SVC(C=1, kernel='linear', gamma='auto'),iris.data, iris.target, cv=5)
```

```python
ker = ['rbf', 'linear']
c = [1,5,10,15,20,25,30]
avg_scores={}

for kval in ker:
    for Cval in c:
        cv_score = cross_val_score(SVC( C=Cval, kernel=kval, gamma='auto'), iris.data, iris.target, cv=5)
        avg_scores[kval+"_"+str(Cval)]=np.average(cv_score)
avg_scores
```

Doing the same as the above for loop easily using GridSearchCV()

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma='auto'),{
    'C':[1,5,10,15,20,25,30],
    'kernel':['rbf','linear']
    }, cv=5, return_train_score=True)

clf.fit(iris.data, iris.target)
clf.cv_results_
```

```python
df1=pd.DataFrame(clf.cv_results_)
df1
```

```python
df1[['param_C','param_kernel','mean_test_score']]
```

Using RandomizedSearchCV()

```python
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(SVC(gamma='auto'), {
        'C': [1,5,10,15,20,25,30],
        'kernel': ['rbf','linear']
    }, 
    cv=5, 
    return_train_score=False, 
    n_iter=3
)
rs.fit(iris.data, iris.target)
pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
```

Supplying the Models as a parameter in the grid Search, using python Dictionaries

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
model_params={
    'svm':{
        'model': SVC(gamma='auto'),
        'params':{
            'C':[1,5,10,15,20,25,30],
            'kernel':['rbf','linear']
        }
    },
    'random_froest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,3,5,7,10]
        }
    },
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,2,4,6,8,10]
        }
    }
}
```

```python
scores=[]

for model_name, mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'], cv=5)
    clf.fit(iris.data,iris.target)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params':clf.best_params_
    })
df =pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df
```

So, SVM is the best classification model for this data with hyperparameters, C=1 and kernel='rbf
